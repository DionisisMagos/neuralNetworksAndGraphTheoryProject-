```python
import tensorflow as tf
import numpy as np
```

```python
# The dataset will be based on the linear equation y = 3x + 1
x_train = np.array([-1, 0, 1, 2, 3, 4], dtype=float)
y_train = np.array([-2, 1, 4, 7, 10, 13], dtype=float)
```

 ------------------------------------------------------------------------------
 we have three distinct layers of vertices: The Input layer, which contains one input node, the x value. The Hidden layer, which contains three neurons and these three neurons receives informations from the input layer. If we had two neurons will be the same result for this equation. Last, the Output layer, which contains one neuron, this one neuron will find the prediction for the value Y (the final output)
 ------------------------------------------------------------------------------

```python
# The hidden layer will have 3 neurons and the output layer will have 1 neuron
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=3, input_shape=[1], name='Hidden_Layer'),
    tf.keras.layers.Dense(units=1, name='Output_Layer')
])
```

--------------------------------------------------------------------------------
 Each edge has a numerical weight, which represents the strength of the connection (adjustable parameters). Input layer => Hidden layer: 1 input node * 3 hidden neurons = 3 weighted edges. Hidden layer => Output layer: 3 hidden neuron * 1 output neuron = 3 weighted edges. Our model's graph will be 3+3=6 edges. The goal pf training is to find the optimal values for these six weights
---------------------------------------------------------------------------------

For each neuron we have that, it receives inputs from the previous layer's nodes along the connected edges. Also, it calculates a sum of these inputs, each multiplied by its corresponding edge's weight. Furthermore, a bias term is added to this weighted sum. Just as the weights, biases are also learnable parameters.

```python
model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
```

```python
# We will print a summary to see the structure
model.summary()
```

```python
# Train the model
print("\nTraining the Model...")
model.fit(x_train, y_train, epochs = 500, verbose=1) # The verbose=0 hides the long training output for a cleaner notebook but because we want to see how the loss changes we will set it to 1
print("Training finished")
```

I observe that the epochs fro more training (like 1000) will have the same loss as 500, because the brain of the computer will find the best loss possible and i could not find a better loss for this equation.

```python
# We will make predictions for a number which the model has ever seen
print("\nTest for a new value which the model has ever seen")
new_x = 20
correct_y = 3 * new_x + 1
prediction_input = np.array([new_x])
prediction = model.predict(prediction_input)

print(f"\nModel prediction for x = {new_x}: {prediction[0][0]}")
print(f"Actual correct answer: {correct_y}")
```

The architecture of the network we have built is a classic example of a Directed Acyclic Graph (DAG), which is the foundational structure for all feedforward neural networks. This can be broken down into two key properties.
The term Directed means that information flows strictly in one direction. In our graph, the computational path is always Input Layer => Hidden Layer => Output Layer. The term Acyclic means that the graph contains no loops or cycles. A signal passes through any given vertex (neuron) only once during a single forward pass. This lack of cycles is the critical feature that defines a feedforward network.
